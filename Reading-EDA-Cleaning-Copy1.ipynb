```python
fig, ax1 = plt.subplots(figsize = (10, 10))
ax2 = ax1.twinx()

#Create by chart on axis 1
bar_container = ax1.bar(counts_by_yr['yr_sold'], counts_by_yr['count'], label = 'Count', color = 'blue')

#Create line chart on axis 2
ax2.plot(counts_by_yr['yr_sold'], counts_by_yr['mean'], color = 'red', label = 'Mean')

#Add labels to bars in bar chart
ax1.bar_label(bar_container, label = counts_by_yr['count'], padding = 5, size = 15, color = 'blue')
    #Added padding so it's a little more separate from dollar amount in 2007

#Change ax2 labels so they are dollar signs
def format_dollars(x, pos):
    return f'${x:,.0f}'
ax2.yaxis.set_major_formatter(ticker.FuncFormatter(format_dollars))

ax1.set_xlabel('Year Sold', size = 15)
ax1.set_ylabel('Count of Houses Sold', color = 'b', size = 15)
ax2.set_ylabel('Average Sales Price', color = 'r', size = 15)

# Initial Labeling - but had some overlap between sale price and houses sold
# #Show Mean sales price label for each year
# for i, (year, mean) in enumerate(zip(counts_by_yr['yr_sold'], counts_by_yr['mean'])):
#     ax2.text(year, mean, f"${mean:,.0f}", color = 'red', ha = 'center', va = 'top', fontsize = 12)

#To help separate the label for sale price and count of houses sold, ChatGPT suggested a bbox 
for year, mean in zip(counts_by_yr['yr_sold'], counts_by_yr['mean']):
    ax2.text(year, mean, f"${mean:,.0f}", color='red', ha='center', va='top', fontsize=12, 
             bbox=dict(facecolor='white', edgecolor='none', pad=3))  # Adjust fontsize and bbox properties
    
    # Add an offset to the text position to move it below the data point
    ax2.text(year, mean, '', va='bottom', fontsize=0, ha='center') 

#Changing x label so there's no decimal points on the year
plt.xticks(ticks = plt.xticks()[0], labels = plt.xticks()[0].astype(int))

plt.title('Number of Houses Sold and Average Price by Year', fontsize = 22)

plt.savefig('../project-2/images/sale_price_sales_by_year.png', dpi = 192)

plt.show();
```

```python
grouped_df = df.groupby(by = ['ms_zoning'], dropna = False) ['saleprice'].agg(['count', 'mean', 'median']).reset_index().sort_values('count', ascending = False)

grouped_df['count_percentage'] = round( ( (grouped_df['count'] / grouped_df['count'].sum() ) * 100), 2)

for col in ['mean', 'median']:
    col_mean = grouped_df[col].mean()
    col_std = grouped_df[col].std()
    grouped_df[f'{col}_zscore'] = (grouped_df[col] - col_mean) / col_std

grouped_df
```
